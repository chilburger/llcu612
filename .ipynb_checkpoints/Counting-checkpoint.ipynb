```python
import os
directory = "austen"
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
import glob
textFiles = glob.glob("austen/*txt")
textFiles
```

```python
totalCharacters = 0
for textFile in textFiles:
    f = open(textFile, "r",encoding="utf-8-sig")
    textString = f.read()
    f.close()
    chars = len(textString)
    print(textFile, "has", chars, "characters")
    totalCharacters += chars
print("total characters: ", totalCharacters)
```

```python
for textFile in textFiles:
    f = open(textFile, "r",encoding="utf-8-sig")
    textString = f.read()
    wordCounter = {}
    with open(textFile,'r',encoding="utf-8-sig") as fh:
      for line in fh:
    # Replacing punctuation and making the string lowercase.
    # The split will spit the line into a list.
        word_list = line.replace(',','').replace('\'','').replace('.','').replace(';','').replace('"','').replace('?','').replace(')','').replace('(','').replace('!','').lower().split()
        for word in word_list:
      # Adding  the word into the wordCounter dictionary.
          if word not in wordCounter:
            wordCounter[word] = 1
          else:
        # if the word is already in the dictionary update its count.
            wordCounter[word] = wordCounter[word] + 1
    f.close()
    print(textFile, "has these words:")
    print('{:15}{:3}'.format('Word','Count'))
    print('-' * 18)
# printing the words and occurrence.
    for  (word,occurance)  in wordCounter.items(): 
      print('{:15}{:3}'.format(word,occurance))
```
