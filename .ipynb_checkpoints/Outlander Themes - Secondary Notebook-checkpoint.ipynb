# Graphing the Themes
The next section of code uses the Natural Langauge Toolkit (NLTK) for text processing. Check out the <a href="https://nbviewer.jupyter.org/github/sgsinclair/alta/blob/master/ipynb/GettingNltk.ipynb">Getting NLTK for Text Processing</a> notebook for a step-by-step guide on installation. We also referred to the <a href="https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xticks.html">Matplotlib</a>, a Python 2D plotting library, to create this graph.<br>

After processing our plain text files, counting the frequencies of each word, converting our string to lowercase, loading our list of stop-words for English and building a frequency list, we then used for-loops to calculate the total count of the theme words for each book. We then graph this information so that comparisons can be made, including how often theme words appear in total for each of the texts.

--still working on this section--


```python
import nltk
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
N = 8
ind = np.arange(N) 
width = 0.8 
numBooks = 1
for textFile in textFiles: #for loop to count words in a text
    f = open(textFile, "r",encoding="utf-8-sig") #opens the text and stores as f
    textString = f.read() #stores text from f in textString
    f.close()
    book = nltk.corpus.gutenberg.words(textFile) # load words
    stopwords = nltk.corpus.stopwords.words("English") # load stopwords
    # filter words that are alphabetic and not in stopword list
    words = [word.lower() for word in book if word[0].isalpha() and not word.lower() in stopwords]
    freqs = nltk.FreqDist(words) # build frequency list
    theme1Total = 0
    theme2Total = 0
    theme3Total = 0
    theme4Total = 0
    theme5Total = 0
    theme6Total = 0
    theme7Total = 0
    theme8Total = 0
    for word in theme1:
        theme1Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme2:
        theme2Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme3:
        theme3Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme4:
        theme4Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme5:
        theme5Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme6:
        theme6Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme7:
        theme7Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme8:
        theme8Total += textString.count(word) #tally the count of theme words after each text file

    counts = [theme8Total, theme7Total, theme6Total, theme5Total, theme4Total, theme3Total, theme2Total, theme1Total]
    if numBooks == 1:
        plt.barh(ind*8, counts, width, align='center', label='Book 1 - Outlander')
    elif numBooks == 2:
        plt.barh(ind*8 - (width), counts, width, align='center',  label='Book 2 - Dragonfly in Amber')
    elif numBooks == 3:
        plt.barh(ind*8 - (width*2), counts, width, align='center', label='Book 3 - Voyager')
    elif numBooks == 4:
        plt.barh(ind*8 - (width*3), counts, width, align='center', label='Book 4 - Drums of Autumn')
    elif numBooks == 5:
        plt.barh(ind*8 - (width*4), counts, width, align='center', label='Book 5 - The Fiery Cross')
    elif numBooks == 6:
        plt.barh(ind*8 - (width*5), counts, width, align='center', label='Book 6 - A Breath of Snow and Ashes')
    elif numBooks == 7:
        plt.barh(ind*8 - (width*6), counts, width, align='center', label='Book 7 - An Echo in the Bone')
    elif numBooks == 8:
        plt.barh(ind*8 - (width*7), counts, width, align='center', label='Book 8 - Written in my own Hearts Blood')
        
    #print("The Love themed words are used", theme1Total, "times in", textFile)
    #print("The Marriage themed words are used", theme2Total, "times in", textFile)
    #print("The Identity themed words are used", theme3Total, "times in", textFile)
    #print("The Family themed words are used", theme4Total, "times in", textFile)
    #print("The Community themed words are used", theme5Total, "times in", textFile)
    #print("The Loyalty themed words are used", theme6Total, "times in", textFile)
    #print("The Nexus themed words are used", theme7Total, "times in", textFile)
    #print("The Forgiveness themed words are used", theme8Total, "times in", textFile)
    #print("The 5 most common words in", textFile, "are: ")
    #print(freqs.most_common(5)) # plot the top 5 words
    print(textFile,"done")

    numBooks = numBooks + 1

plt.yticks(ind * 8, ("Forgiveness", "Nexus", "Loyalty", "Community", "Family", "Identity", "Marriage", "Love"))
plt.xticks(np.arange(0, 3500, step=350))
plt.grid()
plt.legend(loc='best')
plt.xlabel('Occurrences')
plt.title('Outlander Themes')
plt.show()
```
