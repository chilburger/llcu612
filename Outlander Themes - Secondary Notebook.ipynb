```python
import glob
textFiles = glob.glob("outlander/*txt")
textFiles
```

```python
theme1 = ["adore","admire","care","cherish","enamored","fond","long","affection","love","like","embrace","feel","kiss","caress","hold","hug"]
theme2 = ["wedding","matrimony","husband","wife", "spouse", "bride", "companion", "partner", "mate", "Mrs"]
theme3 = ["identity","claire beauchamp","claire randall","sassenach","sorcha","madonna"]
theme4 = ["family","mother","father","brother","sister","aunt","grandmother","grandfather","parent","uncle", "sibiling"]
theme5 = ["community","clan","fellowship","company", "group", "comradery", "friend"]
theme6 = ["loyalty","sacrifice", "allegiance", "devotion", "honesty", "honor"]
theme7 = ["unwavering ","bond","connection"]
theme8 = ["forgive", "pardon"]
```

```python
import nltk
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

for textFile in textFiles: #for loop to count words in a text
    f = open(textFile, "r",encoding="utf-8-sig") #opens the text and stores as f
    textString = f.read() #stores text from f in textString
    f.close()
    book = nltk.corpus.gutenberg.words(textFile) # load words
    stopwords = nltk.corpus.stopwords.words("English") # load stopwords
    # filter words that are alphabetic and not in stopword list
    words = [word.lower() for word in book if word[0].isalpha() and not word.lower() in stopwords]
    freqs = nltk.FreqDist(words) # build frequency list
    theme1Total = 0
    theme2Total = 0
    theme3Total = 0
    theme4Total = 0
    theme5Total = 0
    theme6Total = 0
    theme7Total = 0
    theme8Total = 0
    for word in theme1:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme1Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme2:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme2Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme3:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme3Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme4:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme4Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme5:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme5Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme6:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme6Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme7:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme7Total += textString.count(word) #tally the count of theme words after each text file
    for word in theme8:
        #print("The Love themed words are used", textString.count(word), "times in", textFile)
        theme8Total += textString.count(word) #tally the count of theme words after each text file
        
    print("The Love themed words are used", theme1Total, "times in", textFile)
    print("The Marriage themed words are used", theme2Total, "times in", textFile)
    print("The Identity themed words are used", theme3Total, "times in", textFile)
    print("The Family themed words are used", theme4Total, "times in", textFile)
    print("The Community themed words are used", theme5Total, "times in", textFile)
    print("The Loyalty themed words are used", theme6Total, "times in", textFile)
    print("The Nexus themed words are used", theme7Total, "times in", textFile)
    print("The Forgiveness themed words are used", theme8Total, "times in", textFile)
    print("The 5 most common words in", textFile, "are: ")
    print(freqs.most_common(5)) # plot the top 5 words
    intTheme1 = int(theme1Total)
    intTheme2 = int(theme2Total)
    intTheme3 = int(theme3Total)
    intTheme4 = int(theme4Total)
    intTheme5 = int(theme5Total)
    intTheme6 = int(theme6Total)
    intTheme7 = int(theme7Total)
    intTheme8 = int(theme8Total)
    xThemes = ["Forgiveness", "Nexus", "Loyalty", "Community", "Family", "Identity", "Marriage", "Love"]
    counts = [intTheme8, intTheme7, intTheme6, intTheme5, intTheme4, intTheme3, intTheme2, intTheme1]
    y_pos = np.arange(len(xThemes))
    plt.barh(y_pos, counts, align='center', alpha=0.5)
    plt.yticks(y_pos, xThemes)
    plt.xlabel('Counts')
    plt.title('Theme Word Count')
    plt.show()
    print('-' * 90) #prints the separation dotted line
```
